```python
import matplotlib.pyplot as plt

# Data
mpg_data = [19, 19, 19, 20, 21, 21, 25, 25, 25, 26, 26, 28, 29, 31, 31, 32, 32, 33, 34, 35, 36, 37, 37, 38, 38, 38, 38, 41, 43, 43]

# Sort the data
mpg_data.sort()

# Extract stems and leaves
stems = [int(x/10) for x in mpg_data]
leaves = [x%10 for x in mpg_data]

# Create stem plot
plt.stem(stems, leaves, linefmt='C0-', markerfmt='C0o', basefmt='k-', use_line_collection=True)
plt.xlabel('Stem')
plt.ylabel('Leaf')
plt.title('Stem-and-Leaf Plot')
plt.grid(True)

# Identify outlier
outlier = max(set(mpg_data), key=mpg_data.count)
plt.annotate(f'Outlier: {outlier}', xy=(stems[-1], leaves[-1]), xytext=(-30, 10), textcoords='offset points', arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
grades = [77, 78, 76, 81, 86, 51, 79, 82, 84, 99]

# Sort the data
grades.sort()

# Extract stems and leaves
stems = [int(x/10) for x in grades]
leaves = [x%10 for x in grades]

# Create stem plot
plt.stem(stems, leaves, linefmt='C0-', markerfmt='C0o', basefmt='k-', use_line_collection=True)
plt.xlabel('Stem')
plt.ylabel('Leaf')
plt.title('Stem-and-Leaf Plot for Chemistry Exam Grades')
plt.grid(True)
plt.show()

# Identify potential outliers
Q1 = grades[int(len(grades)*0.25)]
Q3 = grades[int(len(grades)*0.75)]
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [grade for grade in grades if grade < lower_bound or grade > upper_bound]

if outliers:
    print("Potential outliers:")
    print(outliers)
    print("These scores are considered outliers because they fall below Q1 - 1.5*IQR or above Q3 + 1.5*IQR.")
else:
    print("No potential outliers.")
```

```python
import matplotlib.pyplot as plt

# Data
movies = [0, 1, 2, 3, 4]
frequencies = [5, 9, 6, 4, 1]

# Expand the data
expanded_data = []
for i in range(len(movies)):
    expanded_data.extend([movies[i]] * frequencies[i])

# Create box plot
plt.boxplot(expanded_data, vert=False)
plt.xlabel('Number of Movies Watched')
plt.title('Box Plot of Number of Movies Watched')
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Data
years = [7, 14, 15, 18, 19, 20, 22, 23, 26, 40, 42]
frequencies = [1, 3, 1, 1, 4, 3, 1, 1, 1, 2, 2]

# Create expanded data
expanded_data = []
for i in range(len(years)):
    expanded_data.extend([years[i]] * frequencies[i])

# Calculate quartiles
Q1 = np.percentile(expanded_data, 25)
Q3 = np.percentile(expanded_data, 75)

# Calculate Interquartile Range (IQR)
IQR = Q3 - Q1

print("Interquartile Range (IQR):", IQR)
```

```python
import statistics

# Data
years = [7, 14, 15, 18, 19, 20, 22, 23, 26, 40, 42]
frequencies = [1, 3, 1, 1, 4, 3, 1, 1, 1, 2, 2]

# Create expanded data
expanded_data = []
for i in range(len(years)):
    expanded_data.extend([years[i]] * frequencies[i])

# Find the mode
mode = statistics.mode(expanded_data)

print("Mode:", mode)
```

```python
import pandas as pd
import seaborn as sns

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Calculate mean
mean_values = iris.mean()

# Calculate median
median_values = iris.median()

# Calculate mode
mode_values = iris.mode().iloc[0]  # Mode might have multiple values, so we take the first row

# Display the results
print("Measure of Central Tendency for Iris Dataset:")
print("Mean:")
print(mean_values)
print("\nMedian:")
print(median_values)
print("\nMode:")
print(mode_values)
```

```python
import pandas as pd
import seaborn as sns

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Determine outliers
outliers = {}

for column in iris.columns[:-1]:  # Exclude the species column
    Q1 = iris[column].quantile(0.25)
    Q3 = iris[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers[column] = iris[(iris[column] < lower_bound) | (iris[column] > upper_bound)][column]

# Print outliers
print("Outliers:")
for key, value in outliers.items():
    if not value.empty:
        print(key, ":", value.values)
```
