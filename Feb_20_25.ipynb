```python
#Scaling the values of diabetes dataset
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/DACC Lab/diabetes.csv')
X = df.drop(columns=['Outcome'])
y = df['Outcome']

min_max_scaler = MinMaxScaler()
X_min_max_scaled = min_max_scaler.fit_transform(X)

standard_scaler = StandardScaler()
X_standard_scaled = standard_scaler.fit_transform(X)

X_min_max_scaled_df = pd.DataFrame(X_min_max_scaled, columns=X.columns)
X_standard_scaled_df = pd.DataFrame(X_standard_scaled, columns=X.columns)

print("Min-Max Scaled Data:")
print(X_min_max_scaled_df.head())

print("\nStandardized (Z-score) Scaled Data:")
print(X_standard_scaled_df.head())
```

```python
#remove the outliers using

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diabetes dataset
diabetes_data = pd.read_csv('diabetes.csv')  # Replace 'diabetes.csv' with the path to your dataset

# Create a box plot to visualize the distribution of each numerical attribute
plt.figure(figsize=(10, 6))
sns.boxplot(data=diabetes_data)
plt.title('Box Plot of Diabetes Dataset')
plt.xticks(rotation=45)
plt.show()

# Identify outliers using the box plot
Q1 = diabetes_data.quantile(0.25)
Q3 = diabetes_data.quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
diabetes_data_no_outliers = diabetes_data[~((diabetes_data < lower_bound) | (diabetes_data > upper_bound)).any(axis=1)]

# Display the shape of the dataset before and after removing outliers
print("Shape of the original dataset:", diabetes_data.shape)
print("Shape of the dataset after removing outliers:", diabetes_data_no_outliers.shape)
```
